#Errors, Uncertainties, Variance and Covariance

The precision of an experimental result is fundamental to its interpretation.  For example, the value of the parameter $a_\mu$ describing the [anomalous magnetic moment of the muon](http://muon-g-2.fnal.gov/2-the-physics-of-g-2.html) is 
$$
\begin{aligned}
a_\mu &= 11\,659\,208 \times 10^{-10}\mbox{  (experiment)}\\
a_\mu &= 11\,659\,181 \times 10^{-10}\mbox{  (theory)}
\end{aligned}
$$
(as measured by the [Brookhaven E821 experiment](http://www.g-2.bnl.gov/)).

These agree to 6 significant figures &ndash; a triumph of the Standard Model?  Well, **no**, because the uncertainties in these values are $\pm6\times10^{-10}$ and $\pm8\times10^{-10}$ respectively, leading to a discrepancy of $(27\pm10)\times10^{-10}$, or 2.7 error bars.  The probability of a 2.7&sigma; discrepancy arising by chance is only 0.68% (see [notebook on Gaussian errors](gaussian.ipynb)), so this is a hint that the Standard Model prediction may *not* correctly describe this result (and the experiment has consequently moved to Fermilab to redo the measurement with higher sensitivity).

##Terminology

Formally, the **error** in a measurement is the (unknown) difference between the measured value and the true value, whereas the **uncertainty** in a measurement is the (known, or at least estimated) limit to the precision of the measurement caused by limitations of the experimental apparatus (e.g. finite resolution, imperfect calibration, presence of background) or by inherent properties of the quantity being measured (e.g. radioactive decay and other random processes).  However, most physicists talk about "experimental error" when they mean "uncertainty", and "error bars" on a plot are actually a representation of the uncertainty of the measurement, not its (unknown) error.

In many statistical analyses, the key quantity is the **variance** of the measurement.  If a quantity has a true value $x_0$, and an experiment measures this quantity $N$ times with results $x_i$, the variance is the mean squared deviation of $x_i$ from $x_0$:
$$
V_{xx} = \sigma^2 = \frac{1}{N}\sum_{i=1}^{N}(x_i-x_0)^2 
$$
(assuming that $N$ is very large).  The **standard deviation** (or **RMS**, "root-mean-square"), $\sigma$, is the square root of the variance, and represents the probable error of an individual measurement $x_i$. 

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def gaussian(x,mu,sigma):
    return np.exp(-0.5*np.power((x-mu)/sigma,2))/(np.sqrt(2*np.pi)*sigma)

mu = 0.0
sigma=0.5
nbins = 50

x=np.linspace(mu-3*sigma,mu+3*sigma,nbins)

plt.plot(x,gaussian(x,mu,sigma))
plt.errorbar(mu,0.1,xerr=sigma,fmt='bo',ecolor='b')
plt.title('Gaussian distribution and corresponding error bar')
```

##Statistical and systematic errors

**Statistical** (or **random**) errors affect each measured point independently: the mean of a large number of repeated measurements is unaffected, but the scatter about the mean increases as the level of statistical error increases.  An example of a statistical error would be the variation in number of counts per minute observed from a sample of, say, uranium-238 (or any other isotope whose half-life is &gg;1 minute).  

**Systematic** errors affect all measurements in a correlated way (the effect may not be the same on each measurement, but the effect on measurement $i$ is related in a well-defined way to the effect on measurement $j$).  The scatter of the measurements is unaffected, but the mean is (in general) shifted.  An example of a systematic error would be an incorrect calibration of the measuring instrument.

The plot below shows the effect of a 10% statistical error (blue squares) compared to a 10% systematic error (red dots) on a linear relationship.  The gradient is affected by the systematic error, but not by the statistical error.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

m = 2.0
c = 5.0
sigma = 0.1
xmin = 0.0
xmax = 10.0
nbins = 50

x = np.linspace(xmin,xmax,nbins)
y = m*x + c
dystat = np.random.normal(0.0,1.0,nbins)
ydat1 = y + sigma*y*dystat # a 10% random error on y
ydat2 = y + sigma*y        # a 10% systematic error on y

plt.plot(x,y,'b',label='true line')
plt.plot(x,ydat1,'bs',label='10% stat. error')
plt.plot(x,ydat2,'ro',label='10% syst. error')
plt.title('Statistical and systematic errors')
plt.legend(loc='best')
```

Here is another example of a systematic error, this time of the type that might be caused by saturation of your measuring instrument.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

m = 2.0
c = 5.0
xmin = 0.0
xmax = 10.0
ysat = 2.0*(m*xmax+c)
nbins = 50

x = np.linspace(xmin,xmax,nbins)
y = m*x + c
y1 = (m*x + c)/np.sqrt(1+np.power(y/ysat,2))

plt.plot(x,y,'b',label='true line')
plt.plot(x,y1,'ro',label='with saturation')
plt.title('Statistical and systematic errors')
plt.legend(loc='best')
```

In many types of analysis, both statistical and systematic errors are present.  Because they have different effects, they are normally quoted separately in the result, e.g.
$$
H_0 = 71 \pm 2\mbox{ (stat.)} \pm 6\mbox{ (syst.)}\ \mbox{km s}^{-1}\mbox{ Mpc}^{-1}.
$$

##Correlated and uncorrelated errors

When propagating errors using rules such as $\Delta(x+y) = \sqrt{(\Delta x)^2 + (\Delta y)^2}$, we are assuming that $x$ and $y$ are **uncorrelated**: that is, the difference between measurement $x_i$ and true value $x_0$ has no effect on the difference between $y_i$ and $y_0$.  If we were to make a scatter plot of uncorrelated variables, the result should be a blob which is symmetrical about the $x$ and $y$ axes:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

xtrue = 1.0
sigx = 10.0

ytrue = 10.0
sigy = 5.0

xdata=np.random.normal(xtrue,sigx,1000)
ydata=np.random.normal(ytrue,sigy,1000)

plt.plot(xdata,ydata,'b.')
plt.title('Uncorrelated variables')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
```

For **correlated** variables, on the other hand, the difference between $x_i$ and $x_0$ **does** affect the difference between $y_i$ and $y_0$.  For example, the gradient and intercept of a fit to a straight line are correlated:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def line(x,y):    # this is an unweighted least-squares fit
    xmean = x.mean()
    m = ((x-xmean)*y).sum()/((x-xmean)*x).sum()
    c = y.mean()-m*xmean
    return m,c

xmin = 0.0
xmax = 5.0
m = 2.0
c = 1.0
sigy = 0.5

mfit=np.zeros(100)
cfit=np.zeros(100)
xdata=np.linspace(xmin,xmax,20)

for i in range(len(mfit)):  #generate 1000 instances of this relationship and fit each one separately
    ydata=m*xdata+c+np.random.normal(0.0,sigy,20)
    mfit[i],cfit[i] = line(xdata,ydata)
    
plt.plot(mfit,cfit,'b.')    
plt.title('Gradient and intercept of straight line fit')
plt.xlabel('fitted gradient')
plt.ylabel('fitted intercept')
    
mave = mfit.mean()
cave = cfit.mean()
mccov = (mfit-mave)*(cfit-cave)
mccorr = mccov.mean()/(mfit.std()*cfit.std())
print('The correlation coefficient between m and c is %5.3f'%mccorr)
```

Correlation of variables is quantified by the **covariance**
$$
V_{xy} = \left\langle(x_i-x_0)(y_i-y_0)\right\rangle.
$$
The **correlation coefficient** is $V_{xy}/(\sigma_x\sigma_y)$.  It lies between &ndash;1 (perfect anticorrelation) and +1 (perfect correlation).  The correlation coefficient of $m$ and $c$ in the above example shows that the gradient and intercept of the fitted straight line are very strongly anticorrelated.

Below are some examples of different degrees of correlation.  In each case, the variances of $x$ and $y$ are both 1.0.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

mean=[0,0]
cov1=[[1,0.95],[0.95,1]]
x1,y1=np.random.multivariate_normal(mean,cov1,500).T

cov2=[[1,0.5],[0.5,1]]
x2,y2=np.random.multivariate_normal(mean,cov2,500).T

cov3=[[1,-0.5],[-0.5,1]]
x3,y3=np.random.multivariate_normal(mean,cov3,500).T

cov4=[[1,-0.95],[-0.95,1]]
x4,y4=np.random.multivariate_normal(mean,cov4,500).T

plt.figure(1,figsize=(6,6))
plt.subplot(221)
plt.plot(x1,y1,'b.')
plt.axis('equal')

plt.subplot(222)
plt.plot(x2,y2,'r.')
plt.axis('equal')

plt.subplot(223)
plt.plot(x3,y3,'r.')
plt.axis('equal')

plt.subplot(224)
plt.plot(x4,y4,'b.')
plt.axis('equal')
```

###Notes

1. If, as is usually the case, we do not know the true values $(x_0,y_0)$ and instead use the means $(\bar{x},\bar{y})$, then we need to subtract one degree of freedom in calculating the (co)variances:
$$
\begin{aligned}
V_{xx} = \frac{1}{N-1}\sum_{i=1}^N(x_i - \bar{x})^2 &= \frac{N}{N-1}\left(\overline{x^2}-\bar{x}^2\right)\\
V_{xy} = \frac{1}{N-1}\sum_{i=1}^N(x_i-\bar{x})(y_i-\bar{y}) &= \frac{N}{N-1}\left(\overline{xy}-\bar{x}\bar{y}\right) \\
V_{yy} = \frac{1}{N-1}\sum_{i=1}^N(y_i - \bar{y})^2 &= \frac{N}{N-1}\left(\overline{y^2}-\bar{y}^2\right)
\end{aligned}
$$

2. The covariance matrix is always symmetric, $V_{ij} = V_{ji}$.

3. The correlation coefficient is rather a blunt tool: it can give misleading results if the relationship between $x_i-x_0$ and $y_i-y_0$ is more complicated, as in the example below: $x-\bar{x}$ and $y-\bar{y}$ are clearly correlated, but the correlation coefficient is nearly zero because of the symmetry of the distribution (if we remove the symmetry by reversing the sign of $x_i-\bar{x}$ for negative $x_i$, the correlation coefficient is much larger).

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

sigx = 2.0
sigy = 1.0

x=np.random.normal(0.0,sigx,5000)
dy=np.random.normal(0.0,sigy,5000)
y=(x*x+dy)/(sigx*sigx)

xmean=x.mean()
xsig = x.std()
ymean=y.mean()
ysig = y.std()
xycov=(x-xmean)*(y-ymean)
absxycov=np.where(x>0,(x-xmean)*(y-ymean),-(x-xmean)*(y-ymean))
xycorr = xycov.mean()/(xsig*ysig)
absxycorr=absxycov.mean()/(xsig*ysig)
print('Mean x = %5.3f'%xmean,'; sigma(x) = %5.3f'%xsig)
print('Mean y = %5.3f'%ymean,'; sigma(y) = %5.3f'%ysig)
print('Correlation coefficient = %5.3f'%xycorr) 
print('Correlation coefficient for |x| = %5.3f'%absxycorr)

plt.plot(x,y,'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Not really uncorrelated variables!')
```

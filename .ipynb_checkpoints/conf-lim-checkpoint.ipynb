#Confidence intervals, limits and exclusions

If we know the probability density function, it is easy to define a confidence interval around the true population mean &mu; such that there is some well-defined probability (conventionally 68.3%, because of the properties of the Gaussian) that any measured value *x* will fall in that range.  Unfortunately, what we actually have is the measured value *x*, and what we want is information about the true value &mu;.  This is considerably more difficult!

There are two fundamentally different approaches to statistical inference.

1. In **frequentist** statistics, it is assumed that there is an underlying fixed set of true parameter values, and your dataset is one of an infinite number of possible datasets that could have been drawn from the underlying true distribution.  In this picture, a 68.3% confidence interval is defined such that in 68.3% of the possible datasets, the interval would contain the fixed true parameter.

2. In **Bayesian** statistics, the data are fixed, and it is assumed that it is legitimate to define a probability density function *for the true parameter &mu;*.  This pdf represents the experimenter's *degree of belief* in that particular value of &mu;.  In this picture, the equivalent of a confidence interval is called a **credible interval**: a 68.3% credible interval around the (fixed) data has a 68.3% chance of containing the true parameter.  Bayesian statistics is so called because it makes use of **Bayes' theorem** of conditional probability: *P*(*A*|*B*) = *P*(*B*|*A*)*P*(*A*)/*P*(*B*) (the probability of hypothesis *A* being true, given dataset *B* &ndash; which is what we want to know &ndash; is equal to the probability of observing dataset *B* given that hypothesis *A* is true &ndash; which we can calculate &ndash; times our estimate of the probability of *A* being true in the absence of data from the experiment (the **prior probability** of *A*), divided by the probability of observing *B* summed over all possible hypotheses).  *P*(*B*) does not depend on the choice of hypothesis and is just a normalisation factor; the challenge in Bayesian analysis is defining the prior probability *P*(*A*).

Bayesian inference is gaining in popularity as computational tools become available (especially [**Markov Chain Monte Carlos**](www.math.chalmers.se/Stat/Grundutb/CTH/mve250/0809/mcmc1.pdf); there are several implementations in Python, such as [emcee](http://dan.iel.fm/emcee/current/)).  It has the great merit that prior probabilities can be defined to make use of existing knowledge about *A* (e.g. pre-existing measurements, or unphysical regions such as negative mass).  On the other hand, the choice of *P*(*A*) is essentially subjective (it reflects your *degree of belief* in *A* prior to doing the experiment), and it clearly affects the result *P*(*A*|*B*).

##Estimating frequentist confidence intervals

To estimate frequentist confidence intervals, we

* define our confidence interval;
* construct the appropriate confidence interval *around the true value &mu;*, for a range of &mu; which spans the expected range;
* use this to construct a *confidence band* in the &mu;-*x* plane;
* read off the appropriate confidence interval for a given *x*.

There are two types of confidence interval: **central intervals** where the confidence interval has both a lower and an upper bound, and **upper confidence limits** where the lower bound is zero.  For central intervals, it is usual to set the bounds *x*<sub>low</sub>, *x*<sub>high</sub> such that 

$$
P(x<x_{\mathrm{low}}|\mu) = P(x>x_{\mathrm{high}}|\mu) = (1-\alpha)/2,
$$

where &alpha; is the desired confidence level, e.g. 0.683 for a "1&sigma;" error bar.  This distributes the "excluded" probability symmetrically on either side of &mu;.  For an upper limit, we would choose *P*(*x* &lt; *x*<sub>lim</sub>|&mu;) = 1 &minus; &alpha;.

For simple probability distributions, these bounds can be calculated analytically.  For example, if we have a non-negative variable whose pdf is a negative exponential, *P*(*x*) = e<sup>&minus;*x*/&mu;</sup>/&mu;, then *P*(*x* &lt; *x*<sub>low</sub>|&mu;) = 1 &minus; e<sup>&minus;*x*<sub>low</sub>/&mu;</sup> and *P*(*x* &gt; *x*<sub>high</sub>|&mu;) =  e<sup>&minus;*x*<sub>high</sub>/&mu;</sup>.  These can be solved analytically for any value of &alpha;, as shown below.  Note that the confidence intervals for this probability distribution are very asymmetric.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

xobs = 2.0
alpha = 0.683
mumin = 0.1
mumax = 20
mubins = 99

mu = np.linspace(mumin,mumax,mubins)
x1 = -np.log(0.5*(1+alpha))*mu
x2 = -np.log(0.5*(1-alpha))*mu

mu2 = xobs/(-np.log(0.5*(1-alpha)))
mu1 = xobs/(-np.log(0.5*(1+alpha)))

plt.figure(1,figsize=(10,5))
plt.subplot(121)
mp1 = 0.25*(mumin+mumax)
mp2 = 0.75*(mumin+mumax)
print('Plotting probability distributions for values %.3f'%mp1,'and %.3f'%mp2)
x1mp1 = -np.log(0.5*(1+alpha))*mp1
x2mp1 = -np.log(0.5*(1-alpha))*mp1
x1mp2 = -np.log(0.5*(1+alpha))*mp2
x2mp2 = -np.log(0.5*(1-alpha))*mp2
x = np.linspace(mumin,2*mumax,100)
xc1 = np.linspace(x1mp1,x2mp1,100)
xc2 = np.linspace(x1mp2,x2mp2,100)
y1 = np.exp(-x/mp1)/mp1
y2 = np.exp(-x/mp2)/mp2
yc1 = np.exp(-xc1/mp1)/mp1
yc2 = np.exp(-xc2/mp2)/mp2
plt.plot(x,y1,'b')
plt.fill_between(xc1,yc1,facecolor='skyblue',alpha=0.5)
plt.plot(x,y2,'r')
plt.fill_between(xc2,yc2,facecolor='salmon',alpha=0.5)
plt.title('Sample probability distributions',fontsize=16)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.text(15,0.18,'shaded area = 68.3% CI')

plt.subplot(122)
plt.hlines(mu,x1,x2,'skyblue')
plt.plot(mu,mu,'b')
plt.vlines(xobs,mu1,mu2,'r',linewidth=3)
plt.plot(xobs,xobs,'ro')
plt.title('Frequentist central interval',fontsize=16)
plt.xlabel('observed x')
plt.ylabel('$\mu$',rotation=0,fontsize=14)
plt.ylim(mumin,mumax)
```

##Feldman-Cousins confidence intervals

The frequentist prescription above encounters problems when:

* it is not clear ahead of time whether the result of the experiment will be an upper limit or a central interval &ndash; this is a problem because the prescription for *x*<sub>low</sub> above is different from that for *x*<sub>lim</sub>;

* the measurement may lead to an unphysical value of &mu; &ndash; for example, a signal on a Poisson-distributed background where the measured number of events, *n*<sub>obs</sub>, is less than the expected mean background, &mu;<sub>B</sub>.

Particle physicists commonly address these problems by using the [**Feldman-Cousins prescription**](http://arxiv.org/abs/physics/9711021) (GJ Feldman and RD Cousins, *Phys. Rev.* **D57** (1998) 3873-3889).  The Feldman-Cousins prescription is essentially as follows:

1. For the measured value of *x*, find the **physically allowed** value of &mu; that maximises *P*(*x*|&mu;), &mu;<sub>best</sub>.  For example, if &mu; must be non-negative (e.g. it is a mass, or a number of events), then if *x* &lt; 0, &mu;<sub>best</sub> = 0, whereas if *x* &ge; 0, &mu;<sub>best</sub> = *x*.

2. For each value of &mu; calculate the ratio *R* = *P*(*x*|&mu;)/*P*(*x*|&mu;<sub>best</sub>).

3. When defining the confidence interval &alpha; add values of *x* *in decreasing order of R*.  For a discrete distribution such as the Poisson, this means that you start with the highest *R*-value and then steadily include *x*-values in decreasing order of *R* until the total probability is &ge; &alpha; (in the case of discrete distributions, it is not in general possible to ensure that the confidence interval is *exactly* &alpha;, and the convention is to overshoot ("over-cover") rather than undershoot ("under-cover")).  For a continous distribution, you need to find *x*<sub>1</sub>, *x*<sub>2</sub> such that *R*(*x*<sub>1</sub>) = *R*(*x*<sub>2</sub>) and the integral of *P*(*x*) from *x*<sub>1</sub> to *x*<sub>2</sub> is equal to &alpha;.

This technique is a bit fiddly, especially for continuous distributions where it is necessary to solve numerically for *x*<sub>1</sub> and *x*<sub>2</sub> for each value of &mu;, but it has several advantages:

* by construction, it will not return confidence intervals which include unphysical values of &mu;;
* also by construction, it will never under-cover;
* there is a smooth transition from central intervals to upper limits.

Here is an example of confidence intervals for observing number of events *n* (Poisson-distributed) for a signal mean &mu; in the presence of a background with known mean *b*. To extract the confidence interval for observed value *xobs*, read off the intersection of the red line with the blue band.  Note that the blue band has steps because the Poisson function is discrete; also note how it transitions smoothly from a central interval to an upper limit when *xobs* &lt; *b*.

```python
%matplotlib inline

import numpy as np
import math as mt
import matplotlib.pyplot as plt

def poisson(n,mu): #doing it this way to avoid dividing two very large numbers, mu^n and n!
    p=np.exp(-mu)
    if n == 0:
        return p
    else:
        for i in range(1,n+1):
            p = p*mu/i
    return p

nmax = 30
dtype = [('nobs',int),('prob',float),('R',float)]
values = np.zeros(nmax,dtype=dtype) #values is a structured array

b = 4
alpha = 0.683
xobs = 3

mu = np.linspace(0,nmax-15,10*(nmax-15)+1)
nlow = np.zeros(10*(nmax-15)+1)
nhigh = np.zeros(10*(nmax-15)+1)
im = 0

for m in mu:
    for i in range(nmax):
        mubest=max(0,i-b)
        prbest = poisson(i,mubest+b)
        prob = poisson(i,m+b)
        ratio = prob/prbest
        values[i] = (i,prob,ratio)

    sortv = np.sort(values,order='R') #sort values by R
    nlo = m + b
    nhi = m + b
    prtot = 0
    
    for i in range(nmax):
        nn,pr,rat = sortv[nmax-1-i] #I'm sure there's some way to sort in decreasing order but I don't know it!
        if nn < nlo:
            nlo = nn
        elif nn > nhi:
            nhi = nn
        prtot = prtot + pr
        if prtot >= alpha: #add n-values, and probabilities, until prtot >= required confidence level
            break
            
    nlow[im] = nlo
    nhigh[im] = nhi
    im = im+1

print('Mean expected background = ',b)
print('Confidence interval = ',alpha)
plt.figure(1,figsize=(6,4))
plt.ylim(0,nmax-15)
plt.hlines(mu,nlow,nhigh,'skyblue')
plt.vlines(xobs,0,nmax-15,'r')
plt.title('Feldman-Cousins confidence band',fontsize=14)
plt.xlabel('observed n')
plt.ylabel('signal mean $\mu$')
```

##Bayesian credible intervals

To determine Bayesian credible intervals, we calculate *P*(&mu;|*x*<sub>obs</sub>) = *P*(*x*<sub>obs</sub>|&mu;)*P*(&mu;)/*P*(*x*<sub>obs</sub>).  A credible interval corresponding to confidence level &alpha; is then given by

$$
\int_{\mu_1}^{\mu_2} P(\mu |x_{\mathrm{obs}})\,\mathrm{d}\mu = \alpha.
$$

As with frequentist confidence intervals, there are in principle infinitely many intervals \[&mu;<sub>1</sub>,&mu;<sub>2</sub>\] that would satisfy this requirement.  For a central interval we would conventionally choose &mu;<sub>1</sub> such that 

$$
\int_{\mu_{\mathrm{min}}}^{\mu_1} P(\mu |x_{\mathrm{obs}})\,\mathrm{d}\mu = (1-\alpha)/2,
$$

i.e. half the "excluded" probability is below the interval, and half above.

*P*(*x*<sub>obs</sub>) is a normalisation constant, and doesn't have to be determined explicitly (we could ignore it, and normalise everything by the integral of *P*(*x*<sub>obs</sub>|&mu;)*P*(&mu;) from &mu;<sub>min</sub> to &mu;<sub>max</sub>).  The key feature of Bayesian intervals is the need to define *P*(&mu;), the **prior**.  Obviously, the freedom to define the prior does help with some of the issues that arose with frequentist intervals: one can avoid unphysical regions of &mu; by setting *P*(&mu;) = 0 within those regions, for example.  But it may also make it difficult to compare estimates from different experiments &ndash; note that choosing a "uniform" prior in the variable measured by experiment A may not be equivalent to a "uniform" prior in the variable measured by experiment B.

There is a **huge** amount of literature on Bayesian inference, which is becoming increasingly widely used &ndash; if interested, look it up.  (We do not have space to do it justice here.)

##Nuisance parameters

"Nuisance parameters" are parameters you don't know and don't care about: for example, there may be an uncertainty in the background rate, the flux of neutrinos from your beam, the efficiency of your detector, etc.  You need to take account of these, because they affect the probability density function of your observable *x*, but they are not important to your hypothesis test.  

In Bayesian analysis, you can **marginalise** over the nuisance parameters, which basically means that you integrate them out of your pdf: if you have model parameters **a** and nuisance parameters **&nu;**, you integrate your conditional probability over **&nu;**:

$$
P(\mathbf{a}|x) = \int P(\mathbf{a,v}|x)\,\mathrm{d}\mathbf{v}.
$$

This integral is usually not user-friendly, especially if there are several nuisance parameters: this is where the Markov Chain Monte Carlo technique gets wheeled in.

In frequentist analysis, nuisance parameters are, well, more of a nuisance: there isn't such a neat way of dealing with them.  One widely used method is the **profile likelihood ratio**: for each set of parameters **a**, define the set of nuisance parameters **v**<sub>opt</sub> that maximise the likelihood &#8466;(**a**,**v**).  The **profile likelihood** &#8466;<sub>P</sub>(**a**) = &#8466;(**a**,**v**<sub>opt</sub>) is a function of **a** only, because you have defined the values of the nuisance parameters by maximising the likelihood  Then scale this by the likelihood maximised with respect to both **a** and **v**, &#8466;(**a&#770;**,**v&#770;**).  The result is the profile likelihood ratio,

$$
\lambda_{\mathrm{P}}(\mathbf{a}) = \frac{{\cal L}_{\mathrm{P}}(\mathbf{a})}{{\cal L}(\hat{\mathbf{a}},\hat{\mathbf{v}})}.
$$

This has the useful properties that (1) it is approximately independent of the nuisance parameters and (2) the distribution of &minus;2 ln &lambda;<sub>P</sub>(**a**) approaches a &chi;<sup>2</sup> distribution for large datasets.

##Summary

* Error bars on data points and physics results are "confidence intervals". In frequentist statistics, they should be defined such that, were the measurement to be repeated many times, they would contain the true value in X% of the repetitions (where X = 68.3 for 1&sigma; errors, but is typically 90 or 95 for upper or lower limits).

* Calculating the confidence interval correctly is non-trivial in cases where the underlying distribution is not Gaussian, or where there are constraints on the true value, e.g. that it not be negative.  The Feldman-Cousins prescription is a way to deal with such cases.

* In Bayesian statistics, the equivalent of a confidence interval is a credible interval, which should have an X% probability of containing the true parameter.  The result obtained will depend on the assumed prior probability.

* Many problems will have nuisance parameters, which are variables whose true value is not of interest to the experimenter.  Bayesians can marginalise over nuisance parameters; frequentists need to choose test statistics that are minimally sensitive to them, such as the profile likelihood ratio.

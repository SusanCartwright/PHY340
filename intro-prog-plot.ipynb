##Programming in Python

Python programming is reasonably straightforward.  The key points are:

* you do not have to declare variables: it will work out what type they are from what you put in them;
* rather than {} as in C, Python uses indentation to delimit code blocks, as in the example below;
* there are no pointers;
* comments are introduced by # rather than !

Python is an interpreted language (it converts the program into machine-executable code "on the fly", instead of compiling it first), which means it can be quite slow.  If you are going to do heavy number crunching in Python you need to think carefully about how you write your code, as some constructions are much slower than others (for example, **for**-loops are apparently very slow, and can often be replaced by direct array operations, which are much quicker).

**I am not an expert!**  These notebooks are my first foray into Python programming, so they probably are *not* models of style and efficiency.  If you want to write *good* Python programs, Stuart Littlefair would be a good man to ask.

```python
# An example Python script

%matplotlib inline 
#A special "magic" command that makes matplotlib draw graphs inline in the notebook, instead of writing to a file

import numpy as np #this is the equivalent of the include stateement in C
import matplotlib.pyplot as plt

def gaussian(x,mu,sigma):
    return np.exp(-0.5*np.power((x-mu)/sigma,2))/(2.50663*sigma)
#this defines a Gaussian, normalised to unit area, with mean mu and standard deviation sigma
#notice that the code block defining the function starts with the : after the def statement and terminates whenever
#the lines stop being indented

mu = 0.0
sigma = 1.0
s = np.random.normal(mu,sigma,1000)  #s is now an array of 1000 Gaussian-distributed random numbers

count, bins, patches = plt.hist(s,30,normed=True,histtype='stepfilled',facecolor='salmon',alpha=0.5,label='random numbers') 
#plot a histogram of s, with 30 bins, area normalised to 1, with semi-transparent salmon-coloured fill
#the hist function returns an array containing the bin contents (count), an array of bin edges (bins), 
#and an array of "patches", which I haven't quite worked out yet but don't need!
plt.plot(bins,gaussian(bins,mu,sigma),'r--',linewidth=2,label='Gaussian')
#plot the gaussian function defined above, as a red dashed line with double the standard width
plt.legend(loc=0) #put a legend in the "best" location
plt.title('Gaussian random numbers')
plt.xlabel('x')
plt.ylabel('P(x)')
```

##Plotting with matplotlib

The matplotlib package is extremely comprehensive, and has far too many features to document here!  There is a basic tutorial in Chapter 5 of David Pine's [PyMan](http://www.physics.nyu.edu/pine/pymanual/html/chap5/chap5_plot.html), as well as an enormous [gallery of examples](http://matplotlib.org/examples/index.html) and [formal documentation](http://matplotlib.org/api/pyplot_api.html) on the matplotlib site.

Here is an example of a simple 2D plot, which also illustrates [arithmetic operations on arrays](https://scipy-lectures.github.io/intro/numpy/array_object.html).

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

mus = 0.0
sigmas = 1.0
s = np.random.normal(mus,sigmas,1000)

mut = 0.0
sigmat = 0.2
t = np.random.normal(mut,sigmat,1000)

theta = np.pi/6

x = s*np.cos(theta)-t*np.sin(theta)
y = t*np.cos(theta)+s*np.sin(theta)

plt.plot(x,y,'b.') #'b.' specifies blue points - the default is a blue line
plt.title('Correlated errors')
```
